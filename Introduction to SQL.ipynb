### ENGINEERING SALARY WITH THE LOWEST MEDIAN SALARY

```python
pd.read_sql_query("SELECT Major, Major_category, MIN(MEDIAN)\
                   FROM recent_grads\
                   WHERE Major_category = 'Engineering'", conn)
```

### TOTAL NUMBER OF STUDENTS IN THE DATABASE

```python
pd.read_sql_query("SELECT SUM(Total)\
                   FROM recent_grads", conn)
```

### AVERAGE OF STUDENTS PER MAJOR, THE MINIMUM OF THE MEN, THE MAXIMUM OF WOMEN

```python
pd.read_sql("SELECT AVG(Total), MIN(Men), MAX(Women)\
             FROM recent_grads", conn)
```

### THE NUMBER OF MAJORS, THE MAXIMUM VALUE OF UNEMPLOYMENT RATES WITH ALIASES

```python
pd.read_sql_query("SELECT COUNT(*) AS 'Number of Majors', MAX(Unemployment_rate) AS 'Highest Unemployment Rate'\
                   FROM recent_grads", conn)
```

### COUNTING THE NUMBER OF DISTINCT MAJORS, MAJOR CODES, AND MAJOR CATEGORIES

```python
pd.read_sql_query("SELECT COUNT(DISTINCT(Major)) AS unique_majors,\
                          COUNT(DISTINCT(Major_category)) AS unique_major_categories,\
                          COUNT(DISTINCT(Major_code)) AS unique_major_codes\
                   FROM recent_grads", conn)
```
